```python
import pandas as pd
import numpy as np

# jupyter notebook full-width display
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
file = 'lit review summary.csv'
lit_review = pd.read_csv(file, na_values='')
lit_review.columns
```

```python
# Literature Review

for i, row in lit_review.iterrows():
    print('Title:', lit_review.loc[i].Title.title())
    print('Authors:', lit_review.loc[i].Authors.replace(';', ',').rstrip(','))
    if not lit_review.Publication.isna().loc[i]:
        print('Publication:', lit_review.loc[i].Publication.title())
    print('Publication Date:', lit_review.loc[i].Year)
    print()
    print(lit_review.loc[i].Citation)
    print()
```

```python
# References

for i, row in lit_review.iterrows():
    print(lit_review.loc[i].Reference.title())
    print()
```
